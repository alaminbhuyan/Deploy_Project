# Spliting the features and level

```python
X = df.drop(labels=['target'],axis=1)
y = df.target
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=4)
```

```python
x_train.shape,x_test.shape
```

# Features Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x_train_sc
```

```python
x_test_sc
```

# Model Building

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
# support vector classifier
from sklearn.svm import  SVC
```

```python
svc = SVC()
```

```python
svc.fit(x_train_sc,y_train)
```

```python
y_pred_svc = svc.predict(x_test_sc)
accuracy_score(y_test,y_pred_svc)
```

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train_sc,y_train)
y_pred_lr = lr.predict(x_test_sc)
accuracy_score(y_test,y_pred_lr)
```

```python
# k-nearest 
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc = KNeighborsClassifier()
knc.fit(x_train_sc,y_train)
y_pred_knc = knc.predict(x_test_sc)
accuracy_score(y_test,y_pred_knc)
```

```python
# Naive Bayes
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train_sc,y_train)
y_pred_gnb = gnb.predict(x_test_sc)
accuracy_score(y_test,y_pred_gnb)
```

```python
# for raw data
gnb2 = GaussianNB()
gnb2.fit(x_train,y_train)
y_pred_gnb2 = gnb2.predict(x_test)
accuracy_score(y_test,y_pred_gnb2)
```

```python
# Decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train_sc,y_train)
y_pred_dtc = dtc.predict(x_test_sc)
accuracy_score(y_test,y_pred_dtc)
```

```python
# for raw data
dtc2 = DecisionTreeClassifier(criterion="entropy")
dtc2.fit(x_train,y_train)
y_pred_dtc2 = dtc2.predict(x_test)
accuracy_score(y_test,y_pred_dtc2)
```

```python
# Randomforest
from sklearn.ensemble import RandomForestClassifier
```

```python
# for raw data
rfc = DecisionTreeClassifier()
rfc.fit(x_train,y_train)
y_pred_rfc = rfc.predict(x_test)
accuracy_score(y_test,y_pred_rfc)
```

```python
rfc2 = DecisionTreeClassifier()
rfc2.fit(x_train_sc,y_train)
y_pred_rfc2 = rfc2.predict(x_test_sc)
accuracy_score(y_test,y_pred_rfc2)
```

```python
# adaboost classifer
from sklearn.ensemble import AdaBoostClassifier
```

```python
# for raw data
abc = AdaBoostClassifier(base_estimator=RandomForestClassifier(criterion="entropy",random_state=51),
                         n_estimators=2000,random_state=3)
abc.fit(x_train,y_train)
y_pred_abc = abc.predict(x_test)
accuracy_score(y_test,y_pred_abc)
```

```python
abc2 = AdaBoostClassifier(base_estimator=RandomForestClassifier(criterion="entropy",random_state=51),
                         n_estimators=3000,random_state=50)
abc2.fit(x_train_sc,y_train)
y_pred_abc2 = abc2.predict(x_test_sc)
accuracy_score(y_test,y_pred_abc2)
```

```python
# xgboost classifier
from xgboost import XGBClassifier
```

```python
# for raw data
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
y_pred_xgb = xgb.predict(x_test)
accuracy_score(y_test,y_pred_xgb)
```

```python
xgb2 = XGBClassifier()
xgb2.fit(x_train_sc,y_train)
y_pred_xgb2 = xgb2.predict(x_test_sc)
accuracy_score(y_test,y_pred_xgb2)
```

# XGboost parameter tuning using **grid search cv**

```python
params = {
    "learning_rate"    :[0.05,0.10,0.15,0.20,0.25,0.30],
    "max_depth"        :[3,4,5,6,8,10,12,15],
    "min_child_weight" :[1,3,5,7],
    "gamma"            :[0.0,0.1,0.2,0.3,0.4],
    "colsample_bytree" :[0.3,0.4,0.5,0.7]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=xgb,param_grid=params,scoring='roc_auc',n_jobs=-1,verbose=True)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_
```

```python
xgb3 = XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=0.3, gamma=0.0,
              learning_rate=0.3, max_delta_step=0, max_depth=5,
              min_child_weight=1, missing=None, n_estimators=100, n_jobs=1,
              nthread=None, objective='binary:logistic', random_state=0,
              reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
              silent=None, subsample=1, verbosity=1)
```

```python
xgb3.fit(x_train,y_train)
y_pred_xgb3 = xgb3.predict(x_test)
accuracy_score(y_test,y_pred_xgb3)
```
